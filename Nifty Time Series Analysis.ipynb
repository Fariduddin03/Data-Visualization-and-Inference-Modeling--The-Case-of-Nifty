```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df4['Year'] = pd.to_datetime(df4['Year'], format='%Y')


df4.set_index('Year', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df4['Annual'], marker='o', linestyle='-')
plt.title('Annual Stock Market Returns Over Time')
plt.xlabel('Year')
plt.ylabel('Annual Returns (%)')
plt.grid(True)
plt.show()

decomposition = seasonal_decompose(df4['Annual'], model='additive', extrapolate_trend='freq', period=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df4['Annual'], label='Original', marker='o', linestyle='-')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(trend, label='Trend', marker='o', linestyle='-')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(seasonal, label='Seasonal', marker='o', linestyle='-')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residual, label='Residual', marker='o', linestyle='-')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import itertools


p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

best_aic = np.inf
best_params = None
for param in pdq:
    try:
        model = ARIMA(df4['Annual'], order=param)
        results = model.fit()
        if results.aic < best_aic:
            best_aic = results.aic
            best_params = param
    except:
        continue

print("Best ARIMA Parameters:", best_params)

best_model = ARIMA(df4['Annual'], order=best_params)
best_results = best_model.fit()

forecast = best_results.forecast(steps=10)  
print("Forecasted Returns:", forecast)
```
